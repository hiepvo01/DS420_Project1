# Data Preprocessing

## Fetch restaurant, mall, park data

```python
import json

restaurants = []
shoppings = []
parks = []

with open('places.json') as f:
    data = json.load(f)
for d in data: 
    key = list(d.keys())[0]
    if key == 'restaurant':
        restaurants = d[key]['coors']
    elif key == 'park':
        parks = d[key]['coors']
    else:
        shoppings = d[key]['coors']
    print(key)
    print(d[key]['coors'])
```

```python
# Filter duplicate coordinates in shopping
shoppings = [x for x in shoppings if x not in restaurants and x not in parks]
# There are 55 remaining malls
len(shoppings)
```

```python
len(parks)
```

## Add distance to nearest place to data

```python
from math import cos, asin, sqrt, pi

def distance(lat1, lon1, lat2, lon2):
    p = pi/180
    a = 0.5 - cos((lat2-lat1)*p)/2 + cos(lat1*p) * cos(lat2*p) * (1-cos((lon2-lon1)*p))/2
    return 12742 * asin(sqrt(a))
```

```python
print(distance(47.5112, -122.257, 47.6101497, -122.2015159))
```

```python
df = pd.read_csv("../kc_house_data.csv")
df.head()
```

```python
def getCountDist(r):
    counts = []
    mins = []
    for idx in range(len(df)):
        count = 0
        min_dis =1000
        lat1 = df.iloc[idx].lat
        long1 = df.iloc[idx].long
        for res in r:
            d = distance(lat1, long1, res[0], res[1])
            if d < 2:
                count += 1
            if d < min_dis:
                min_dis = d
        counts.append(count)
        mins.append(min_dis)
    return counts, mins
```

```python
counts, mins = getCountDist(restaurants)
df["Restaurant_Count"] = counts
df["Restaurant_Distance"] = mins

counts, mins = getCountDist(shoppings)
df["Shopping_Count"] = counts
df["Shopping_Distance"] = mins

counts, mins = getCountDist(parks)
df["Park_Count"] = counts
df["Park_Distance"] = mins
```

## Add price category column for Classifier model

```python
price_cat = []
for price in df.price:
    if price < 321950:
        price_cat.append(1)
    elif(price < 450000):
        price_cat.append(2)
    elif(price < 645000):
        price_cat.append(3)
    else:
        price_cat.append(4)
df['price_category'] = price_cat
```

# Model Application

## Tree Classifier vs Tree Regressor

```python
# Question 7 - Prepare Model
# Linear Regression
# Data prepartion for model - Put number of bikes as Y and other attributes as Xs
X = train_set.drop("price_category", axis=1) 
y = train_set["price_category"].copy()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score 
from sklearn.model_selection import train_test_split
tree_clf = DecisionTreeClassifier(max_depth=4, random_state=42, criterion='gini')
tree_clf.fit(X, y)
```

```python
def display_scores(scores):  
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())
```

```python
#forest_rmse
scores = cross_val_score(tree_clf, X, y,scoring="neg_mean_squared_error", cv=10) 
tree_rmse_scores = np.sqrt(-scores)
display_scores(tree_rmse_scores)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree_clf = DecisionTreeClassifier(max_depth=4, random_state=42, criterion='gini')
tree_clf.fit(X, y)
```

```python
#forest_rmse
scores = cross_val_score(tree_clf, X, y,scoring="neg_mean_squared_error", cv=10) 
tree_rmse_scores = np.sqrt(-scores)
display_scores(tree_rmse_scores)
```

```python
from sklearn.tree import DecisionTreeRegressor
tree_reg = DecisionTreeRegressor(max_depth=4, random_state=42, criterion='mse')
tree_reg.fit(X, y) 
#forest_rmse
scores = cross_val_score(tree_reg, X, y,scoring="neg_mean_squared_error", cv=10) 
tree_rmse_scores = np.sqrt(-scores)
display_scores(tree_rmse_scores)
```
